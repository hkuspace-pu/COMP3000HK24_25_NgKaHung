```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import numpy as np

file_path = './workloads/workloads.csv' 
data = pd.read_csv(file_path)

time_data = data.iloc[:, 5] 
time_data = time_data.dropna() 

mean = time_data.mean()
std_dev = time_data.std()
print(f"Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}")
```

```python
plt.figure(figsize=(8, 6))

sns.kdeplot(time_data, color='blue', linewidth=2, label='KDE Plot')

sns.histplot(time_data, bins=10, kde=False, color='skyblue', alpha=0.4, label='Histogram',stat="density")

plt.title('KDE Plot with Histogram', fontsize=16)
plt.xlabel('Task Time (seconds)', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.violinplot(x=time_data, color='skyblue', inner='quartile', linewidth=1.5)

plt.title('Violin Plot', fontsize=16)
plt.xlabel('Task Time (seconds)', fontsize=14)
plt.grid(alpha=0.3)

plt.show()
```

```python
import numpy as np

bin_edges = np.histogram_bin_edges(time_data, bins=10)
bin_width = bin_edges[1] - bin_edges[0]

hist, _ = np.histogram(time_data, bins=10, density=True)
total_area = np.sum(hist * bin_width)
print("Total area of histogram:", total_area)
```
