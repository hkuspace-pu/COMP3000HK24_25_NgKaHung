```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    "ML-KEM-512": {"Public Key": 800, "Private Key": 1632, "Ciphertext/Signature": 768},
    "ML-KEM-768": {"Public Key": 1184, "Private Key": 2400, "Ciphertext/Signature": 1068},
    "ML-KEM-1024": {"Public Key": 1568, "Private Key": 3168, "Ciphertext/Signature": 1568},
    "Falcon": {"Public Key": 897, "Private Key": 1281, "Ciphertext/Signature": 690},
    "SPHINCS+": {"Public Key": 64, "Private Key": 128, "Ciphertext/Signature": 29792},
    "Dilithium": {"Public Key": 1312, "Private Key": 2528, "Ciphertext/Signature": 2420},
    "RSA (2048)": {"Public Key": 256, "Private Key": 2048, "Ciphertext/Signature": 256},
    "DSA (2048)": {"Public Key": 256, "Private Key": 256, "Ciphertext/Signature": 256},
    "ECDSA (P-256)": {"Public Key": 64, "Private Key": 32, "Ciphertext/Signature": 64}
}

algorithms = list(data.keys())
public_key_sizes = [data[algo]["Public Key"] for algo in algorithms]
private_key_sizes = [data[algo]["Private Key"] for algo in algorithms]
ciphertext_sizes = [data[algo]["Ciphertext/Signature"] for algo in algorithms]

bar_width = 0.25
y = np.arange(len(algorithms))

plt.figure(figsize=(12, 8))

plt.barh(y - bar_width, public_key_sizes, height=bar_width, label='Public Key Size (bytes)', color='skyblue')
plt.barh(y, private_key_sizes, height=bar_width, label='Private Key Size (bytes)', color='salmon')
plt.barh(y + bar_width, ciphertext_sizes, height=bar_width, label='Ciphertext/Signature Size (bytes)', color='lightgreen')

plt.xscale('log')

plt.ylabel('Algorithms', fontsize=12)
plt.xlabel('Size (Log Scale, in Bytes)', fontsize=12)
plt.title('Comparison of Key and Ciphertext Sizes for Various Cryptographic Algorithms (Log Scale)', fontsize=14)
plt.yticks(y, algorithms, fontsize=10)
plt.legend(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
