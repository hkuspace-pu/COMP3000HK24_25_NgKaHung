```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob

sns.set_theme(style="whitegrid")

file_pattern = "PQC_Performance_Results_*.csv" 
files = glob.glob(file_pattern)

all_data = []

for file in files:
    df = pd.read_csv(file)
    
    filename = file.split("/")[-1] 
    parts = filename.replace("PQC_Performance_Results_", "").replace(".csv", "").split("_")
    browser = parts[0] 
    environment = parts[1]
    
    df["Browser"] = browser.capitalize() 
    df["Environment"] = environment.capitalize() 
    
    all_data.append(df)

df = pd.concat(all_data, ignore_index=True)

df["Mean Time (ms)"] = pd.to_numeric(df["Mean Time (ms)"], errors="coerce")

webassembly_data = df[df["Environment"] == "Webassembly"]

print(webassembly_data["Operation"].unique())
print(webassembly_data["Browser"].unique())

palette = {
    "Chrome": "#FFBE7A",
    "Edge": "#82B0D2",
    "Firefox": "#FA7F6F"
}
markers = {
    "keygen": "^",    
    "encapsulate": "s", 
    "decapsulate": "o" 
}

plt.figure(figsize=(16, 10))

sns.scatterplot(
    data=webassembly_data,
    x="Algorithm",    
    y="Mean Time (ms)",     
    hue="Browser",          
    style="Operation",
    markers=markers,        
    palette=palette,     
    sizes=200,  
    alpha=0.8    
)

plt.yscale("log")

plt.title("PQC Algorithm Performance (WebAssembly, Log Scale)", fontsize=16)
plt.xlabel("Algorithm", fontsize=14)
plt.ylabel("Mean Time (ms) [Log Scale]", fontsize=14)

plt.xticks(rotation=45, ha="right")

handles, labels = plt.gca().get_legend_handles_labels()

new_labels = []
for label in labels:
    if label == "encapsulate":
        new_labels.append("Encapsulation/Signing")
    elif label == "decapsulate":
        new_labels.append("Decapsulation/Verification")
    else:
        new_labels.append(label)

plt.legend(handles, new_labels, title="Legend", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

file_name = "PQC_Performance_Results_chrome_webassembly_laptop.csv"

df = pd.read_csv(file_name)

df["Environment"] = "WebAssembly"

df["Mean Time (ms)"] = pd.to_numeric(df["Mean Time (ms)"], errors="coerce")

print(df["Operation"].unique())

operation_palette = {
    "keygen": "#FFBE7A",          # Key Generation
    "encapsulate": "#82B0D2",     # Encapsulation/Signing
    "decapsulate": "#FA7F6F"      # Decapsulation/Verification
}

markers = {
    "keygen": "^",    
    "encapsulate": "s", 
    "decapsulate": "o" 
}

plt.figure(figsize=(16, 10))

for operation in ["encapsulate", "decapsulate", "keygen"]:
    sns.scatterplot(
        data=df[df["Operation"] == operation],
        x="Algorithm",
        y="Mean Time (ms)",
        hue="Operation", 
        style="Operation",
        markers={operation: markers[operation]}, 
        palette={operation: operation_palette[operation]}, 
        s=300, 
        alpha=0.8,
        legend=False  # 
    )

from matplotlib.lines import Line2D

legend_elements = [
    Line2D([0], [0], marker="^", color='w', markerfacecolor="#FFBE7A", markersize=10, label="Key Generation"),
    Line2D([0], [0], marker="s", color='w', markerfacecolor="#82B0D2", markersize=10, label="Encapsulation/Signing"),
    Line2D([0], [0], marker="o", color='w', markerfacecolor="#FA7F6F", markersize=10, label="Decapsulation/Verification")
]

plt.legend(handles=legend_elements, title="Operation", fontsize=12, title_fontsize=12)

plt.yscale("log")

plt.title("PQC Algorithm Performance (WebAssembly, Laptop, Log Scale)", fontsize=16)
plt.xlabel("Algorithm", fontsize=14)
plt.ylabel("Mean Time (ms) [Log Scale]", fontsize=14)

plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```
