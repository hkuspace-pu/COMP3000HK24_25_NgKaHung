```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob

sns.set_theme(style="whitegrid")

file_pattern = "PQC_Performance_Results_*.csv" 
files = glob.glob(file_pattern)

all_data = []

for file in files:
    df = pd.read_csv(file)
    
    filename = file.split("/")[-1] 
    parts = filename.replace("PQC_Performance_Results_", "").replace(".csv", "").split("_")
    browser = parts[0] 
    environment = parts[1]
    
    df["Browser"] = browser.capitalize() 
    df["Environment"] = environment.capitalize() 
    
    all_data.append(df)

df = pd.concat(all_data, ignore_index=True)

df["Mean Time (ms)"] = pd.to_numeric(df["Mean Time (ms)"], errors="coerce")

webassembly_data = df[df["Environment"] == "Webassembly"]

print(webassembly_data["Operation"].unique())
print(webassembly_data["Browser"].unique())

palette = {
    "Chrome": "#FFBE7A",
    "Edge": "#82B0D2",
    "Firefox": "#FA7F6F"
}
markers = {
    "keygen": "^",    
    "encapsulate": "s", 
    "decapsulate": "o" 
}

plt.figure(figsize=(16, 10))

sns.scatterplot(
    data=webassembly_data,
    x="Algorithm",    
    y="Mean Time (ms)",     
    hue="Browser",          
    style="Operation",
    markers=markers,        
    palette=palette,     
    sizes=200,  
    alpha=0.8    
)

plt.yscale("log")

plt.title("PQC Algorithm Performance (WebAssembly, Log Scale)", fontsize=16)
plt.xlabel("Algorithm", fontsize=14)
plt.ylabel("Mean Time (ms) [Log Scale]", fontsize=14)

plt.xticks(rotation=45, ha="right")

handles, labels = plt.gca().get_legend_handles_labels()

new_labels = []
for label in labels:
    if label == "encapsulate":
        new_labels.append("Encapsulation/Signing")
    elif label == "decapsulate":
        new_labels.append("Decapsulation/Verification")
    else:
        new_labels.append(label)

plt.legend(handles, new_labels, title="Legend", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
df = pd.concat(all_data, ignore_index=True)

df["Mean Time (ms)"] = pd.to_numeric(df["Mean Time (ms)"], errors="coerce")

plt.figure(figsize=(16, 10))

markers = {
    "keygen": "^",  
    "encapsulate": "s", 
    "decapsulate": "o"
}
palette = {
    "Webassembly": "blue",
    "Javascript": "orange"
}


sns.scatterplot(
    data=df,
    x="Algorithm",  
    y="Mean Time (ms)",  
    hue="Environment",
    style="Operation", 
    size="Browser",   
    markers=markers, 
    palette=palette,
    alpha=0.8   
)

plt.yscale("log")

plt.title("PQC Algorithm Performance: WebAssembly vs JavaScript (Log Scale)", fontsize=16)
plt.xlabel("Algorithm", fontsize=14)
plt.ylabel("Mean Time (ms) [Log Scale]", fontsize=14)

plt.xticks(rotation=45, ha="right")

handles, labels = plt.gca().get_legend_handles_labels()

new_handles = []
new_labels = [] 

environment_legend = {
    "Webassembly": "WebAssembly",
    "Javascript": "JavaScript"
}

operation_legend = {
    "keygen": "Key Generation",
    "encapsulate": "Encapsulation/Signing",
    "decapsulate": "Decapsulation/Verification"
}

browser_legend = {
    "Chrome": "Chrome",
    "Edge": "Edge",
    "Firefox": "Firefox"
}

for h, l in zip(handles, labels):
    if l in environment_legend:
        new_handles.append(h)
        new_labels.append(environment_legend[l])
    elif l in operation_legend: 
        new_handles.append(h)
        new_labels.append(operation_legend[l])

for browser, color in browser_legend.items():
    new_handles.append(plt.Line2D([0], [0], color="black", marker="o", markersize=10, label=browser))

plt.legend(new_handles, new_labels, title="Legend", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

file_name = "PQC_Performance_Results_chrome_webassembly_laptop.csv"

df = pd.read_csv(file_name)

df["Environment"] = "WebAssembly" 

df["Mean Time (ms)"] = pd.to_numeric(df["Mean Time (ms)"], errors="coerce")

print(df["Operation"].unique())

operation_palette = {
    "keygen": "#FFBE7A",          # Key Generation
    "encapsulate": "#82B0D2",     # Encapsulation/Signing
    "decapsulate": "#FA7F6F"      # Decapsulation/Verification
}

markers = {
    "keygen": "^",    
    "encapsulate": "s", 
    "decapsulate": "o" 
}


plt.figure(figsize=(16, 10))
df["Operation"] = pd.Categorical(df["Operation"], categories=["encapsulate", "decapsulate", "keygen"], ordered=True)
df = df.sort_values("Operation")

sns.scatterplot(
    data=df,
    x="Algorithm",    
    y="Mean Time (ms)",     
    hue="Operation",      
    style="Operation",   
    markers=markers,        
    palette=operation_palette,
    s=300,  
    alpha=0.8    
)

plt.yscale("log")

plt.title("PQC Algorithm Performance (WebAssembly, Laptop, Log Scale)", fontsize=16)
plt.xlabel("Algorithm", fontsize=14)
plt.ylabel("Mean Time (ms) [Log Scale]", fontsize=14)

plt.xticks(rotation=45, ha="right")

handles, labels = plt.gca().get_legend_handles_labels()

new_labels = []
for label in labels:
    if label == "encapsulate":
        new_labels.append("Encapsulation/Signing")
    elif label == "decapsulate":
        new_labels.append("Decapsulation/Verification")
    elif label == "keygen":
        new_labels.append("Key Generation")
    else:
        new_labels.append(label)

plt.legend(handles, new_labels, title="Operation", fontsize=12)

plt.tight_layout()
plt.show()
```
